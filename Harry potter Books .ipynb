### first i will import the needed libraries

```python

import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Books.csv')
data
```

```python
data.describe()
```

```python
data.info()
```

### I will drop the duplicates if any

```python
data=data.drop_duplicates()
data
```

### I need to check the amount of null values available in the dataset

```python
data.isnull().sum()
```

### i will drop the NA values and then print to make sure there are no more nulls

```python
data=data.dropna()
data.isnull().sum()
```

### Filter the dataset for J.K. Rowling books (author of harrypotter)

```python
jkr = data[data['authors'].str.contains('J.K. Rowling')]

# to show all columns
pd.set_option('display.max_columns', None)       

jkr
```

#### after printing i figures that J.K. Rowling has other books than harry potter so i needed to filter them out again

```python
harrypotter = jkr[jkr['title'].str.contains('Harry Potter and')]
#when i tried with 'Harry Potter' only ,it include  the box sets and collections not just the books so instead i used 'Harry Potter and'
harrypotter
```

```python
harrypotter.info()
```

### calculating the AVG ratings

```python
avg_ratings = harrypotter[['title','average_rating']]
avg_ratings
```

```python
#display the average of each book for visualization
plt.bar(avg_ratings.title,avg_ratings.average_rating)
plt.xticks(rotation=30, ha='right')
# I added thes previous line because the name was too long to be displayed horizontaly on x axis 
plt.ylim((4, 4.7))
```

## 2 ways to calculate the Average rating of the Harry Potter books

```python
avg_r=avg_ratings['average_rating'].sum()/avg_ratings['average_rating'].count()
avg_r
```

```python
#another way to calculate the avg 
weighted_avg = (harrypotter.average_rating * harrypotter.ratings_count).sum() / harrypotter.ratings_count.sum()
weighted_avg
```

```python
#display the average of each book for visualization
plt.bar(avg_ratings.title,avg_ratings.average_rating,color='grey')
plt.xticks(rotation=45, ha='right')
# I added thes previous line because the name was too long to be displayed horizontaly on x axis 
plt.ylim((4, 4.7))
plt.hlines(weighted_avg, xmin=-0.4, xmax=6.4, color='pink')
plt.hlines(avg_r, xmin=-0.4, xmax=6.4, color='cyan')
```
